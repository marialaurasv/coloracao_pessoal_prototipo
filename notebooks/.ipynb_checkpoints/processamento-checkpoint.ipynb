```python
import os
import cv2
import numpy as np
from tqdm import tqdm
from mtcnn import MTCNN
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import shutil

# Caminhos
ROOT = ".."  # Pois o notebook estÃ¡ em notebooks/
PATH_ORIGINAIS = os.path.join(ROOT, "imagens_originais")
PATH_AUMENTADAS = os.path.join(ROOT, "imagens_aumentadas")
PATH_SEGMENTADAS = os.path.join(ROOT, "imagens_segmentadas")
PATH_RESULTADO = os.path.join(ROOT, "cartelas_resultado")
```

```python
from PIL import Image

def aplicar_augmentacoes_basicas(imagem):
    augmentadas = []
    augmentadas.append(imagem)
    augmentadas.append(imagem.transpose(Image.FLIP_LEFT_RIGHT))
    return augmentadas

# Gerar imagens aumentadas se ainda nÃ£o existirem
os.makedirs(PATH_AUMENTADAS, exist_ok=True)
arquivos_existentes = set(os.listdir(PATH_AUMENTADAS))

print("ðŸ”„ Gerando imagens aumentadas (se necessÃ¡rio)...")
for filename in tqdm(sorted(os.listdir(PATH_ORIGINAIS))):
    base_name = os.path.splitext(filename)[0]
    needs_augmentation = any(
        f"{base_name}_aug{i}.jpg" not in arquivos_existentes for i in range(5)
    )

    if not needs_augmentation:
        continue  # jÃ¡ foi aumentado antes

    img_path = os.path.join(PATH_ORIGINAIS, filename)
    try:
        image = Image.open(img_path).convert("RGB")
    except:
        continue

    augmented_images = aplicar_augmentacoes_basicas(image)

    for i, aug_img in enumerate(augmented_images):
        aug_filename = f"{base_name}_aug{i}.jpg"
        aug_path = os.path.join(PATH_AUMENTADAS, aug_filename)
        aug_img.save(aug_path)
```

```python
import os
import cv2
import numpy as np
from PIL import Image
from mtcnn import MTCNN

# Modo de segmentaÃ§Ã£o: 'pele' ou 'rosto'
modo_segmentacao = 'pele'  # Altere para 'rosto' se quiser o modo sem remoÃ§Ã£o de olhos, boca etc.

# Caminhos
entrada_dir = "../imagens_aumentadas/"
saida_dir = "../imagens_segmentadas/" if modo_segmentacao == 'pele' else "../rostos_segmentados/"
os.makedirs(saida_dir, exist_ok=True)

# Detector facial
detector = MTCNN()
```

```python
# ----------------------------
# 2. FUNÃ‡ÃƒO: SEGMENTAR SOMENTE A PELE
# ----------------------------

def detectar_rosto(imagem):
    resultado = detector.detect_faces(imagem)
    if resultado:
        x, y, w, h = resultado[0]['box']
        x, y = max(0, x), max(0, y)
        return imagem[y:y+h, x:x+w]
    else:
        return None

def segmentar_pele_mtcnn(imagem_bgr):
    imagem_rgb = cv2.cvtColor(imagem_bgr, cv2.COLOR_BGR2RGB)
    rosto = detectar_rosto(imagem_rgb)
    if rosto is None:
        print("Rosto nÃ£o detectado.")
        return None

    hsv = cv2.cvtColor(rosto, cv2.COLOR_RGB2HSV)

    # Faixa para pele clara
    lower1 = np.array([0, 20, 80], dtype=np.uint8)
    upper1 = np.array([25, 150, 255], dtype=np.uint8)

    # Faixa para pele mais escura/marrom
    lower2 = np.array([25, 40, 40], dtype=np.uint8)
    upper2 = np.array([50, 150, 255], dtype=np.uint8)

    mask1 = cv2.inRange(hsv, lower1, upper1)
    mask2 = cv2.inRange(hsv, lower2, upper2)

    # Combina mÃ¡scaras
    mask = cv2.bitwise_or(mask1, mask2)

    # OperaÃ§Ãµes morfolÃ³gicas para limpar a mÃ¡scara
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)  # fecha buracos
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)   # remove ruÃ­dos

    # SuavizaÃ§Ã£o
    mask = cv2.GaussianBlur(mask, (5, 5), 0)

    resultado = cv2.bitwise_and(rosto, rosto, mask=mask)
    resultado_bgr = cv2.cvtColor(resultado, cv2.COLOR_RGB2BGR)

    return resultado_bgr
```

```python
# ----------------------------
# 3. FUNÃ‡ÃƒO: SEGMENTAR ROSTO COMPLETO (SEM REMOVER OLHOS, BOCA, ETC)
# ----------------------------

def segmentar_rosto_completo(imagem_pil):
    imagem_np = np.array(imagem_pil)
    resultados = detector.detect_faces(imagem_np)

    if resultados:
        x, y, w, h = resultados[0]['box']
        x, y = max(0, x), max(0, y)
        rosto = imagem_pil.crop((x, y, x + w, y + h))
        return rosto
    else:
        return None
```

```python
# ----------------------------
# 4. LOOP FINAL DE PROCESSAMENTO
# ----------------------------

for nome_arquivo in os.listdir(entrada_dir):
    if not nome_arquivo.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    caminho_entrada = os.path.join(entrada_dir, nome_arquivo)

    if modo_segmentacao == 'pele':
        imagem_bgr = cv2.imread(caminho_entrada)
        if imagem_bgr is None:
            print(f"[ERRO] Falha ao ler: {nome_arquivo}")
            continue

        segmentada = segmentar_pele_mtcnn(imagem_bgr)

        if segmentada is not None:
            caminho_saida = os.path.join(saida_dir, nome_arquivo)
            cv2.imwrite(caminho_saida, segmentada)
            print(f"[OK - PELE] {nome_arquivo}")
        else:
            print(f"[AVISO - PELE] Rosto nÃ£o detectado: {nome_arquivo}")

    else:  # modo_segmentacao == 'rosto'
        imagem = Image.open(caminho_entrada).convert("RGB")
        segmentada = segmentar_rosto_completo(imagem)

        if segmentada is not None:
            caminho_saida = os.path.join(saida_dir, nome_arquivo)
            segmentada.save(caminho_saida)
            print(f"[OK - ROSTO] {nome_arquivo}")
        else:
            print(f"[AVISO - ROSTO] Rosto nÃ£o detectado: {nome_arquivo}")
```

```python
def extrair_features_lab(imagem_bgr):
    imagem_lab = cv2.cvtColor(imagem_bgr, cv2.COLOR_BGR2LAB)
    h, w, _ = imagem_lab.shape
    imagem_flat = imagem_lab.reshape((h * w, 3))

    # Remove pixels pretos (sem pele)
    imagem_flat = imagem_flat[np.any(imagem_flat != [0, 128, 128], axis=1)]

    if imagem_flat.shape[0] == 0:
        return np.zeros(27)

    hist_l = cv2.calcHist([imagem_lab], [0], None, [9], [0, 256]).flatten()
    hist_a = cv2.calcHist([imagem_lab], [1], None, [9], [0, 256]).flatten()
    hist_b = cv2.calcHist([imagem_lab], [2], None, [9], [0, 256]).flatten()

    hist = np.concatenate([hist_l, hist_a, hist_b])
    hist = hist / np.sum(hist)  # normaliza

    return hist
```

```python
from sklearn.cluster import KMeans

def extrair_features_rosto(imagem_bgr):
    h, w, _ = imagem_bgr.shape
    imagem_flat = imagem_bgr.reshape((h * w, 3))

    # Remove fundo preto
    imagem_flat = imagem_flat[np.any(imagem_flat != [0, 0, 0], axis=1)]
    if imagem_flat.shape[0] == 0:
        return np.zeros(9)

    # Tom mÃ©dio
    media_rgb = np.mean(imagem_flat, axis=0)

    # Contraste (desvio padrÃ£o da luminÃ¢ncia)
    imagem_gray = cv2.cvtColor(imagem_bgr, cv2.COLOR_BGR2GRAY)
    contraste = np.std(imagem_gray)

    # Cor dominante com KMeans
    kmeans = KMeans(n_clusters=1, random_state=42, n_init='auto')
    cor_dominante = kmeans.fit(imagem_flat).cluster_centers_[0]

    # Junta tudo (tom mÃ©dio, contraste e cor dominante)
    features = np.concatenate([
        media_rgb,                  # 3 valores
        [contraste],                # 1 valor
        cor_dominante               # 3 valores
    ])

    return features  # total: 7
```

```python
import pandas as pd
from tqdm import tqdm

PASTA_ENTRADA = "../rostos_segmentados/" if modo_segmentacao == 'rosto' else "../imagens_segmentadas/"
CSV_SAIDA = "../features_rosto.csv" if modo_segmentacao == 'rosto' else "../features_pele.csv"

features = []
nomes = []

print(f"ðŸ“¥ Extraindo caracterÃ­sticas de {PASTA_ENTRADA}...")
for nome_arquivo in tqdm(os.listdir(PASTA_ENTRADA)):
    caminho = os.path.join(PASTA_ENTRADA, nome_arquivo)
    imagem = cv2.imread(caminho)

    if imagem is None:
        continue

    if modo_segmentacao == 'pele':
        feat = extrair_features_lab(imagem)
    else:
        feat = extrair_features_rosto(imagem)

    features.append(feat)
    nomes.append(nome_arquivo)

# Cria DataFrame
df = pd.DataFrame(features)
df.insert(0, "nome_arquivo", nomes)

# Salva CSV
df.to_csv(CSV_SAIDA, index=False)
print(f"âœ… Features salvas em {CSV_SAIDA}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import umap
import os
import shutil
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import joblib

def processar_e_plotar_salvar(csv_path, titulo, subplot_index, pasta_segmentadas, pasta_resultado, pasta_modelos, prefixo_modelo):
    # Carrega CSV com features
    df = pd.read_csv(csv_path)
    nomes = df['nome_arquivo'].values
    X = df.drop(columns=['nome_arquivo']).values

    # Escalonamento
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # UMAP
    reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
    embedding = reducer.fit_transform(X_scaled)

    # KMeans
    kmeans = KMeans(n_clusters=12, random_state=42, n_init='auto')
    labels = kmeans.fit_predict(embedding)

    # Plot
    cores = plt.cm.tab20(np.linspace(0, 1, 12))
    plt.subplot(1, 2, subplot_index)
    for i in range(12):
        idx = labels == i
        plt.scatter(
            embedding[idx, 0],
            embedding[idx, 1],
            color=cores[i],
            label=f'Cartela {i}',
            s=30, alpha=0.8
        )
    plt.title(titulo, fontsize=14)
    plt.xlabel("UMAP 1")
    plt.ylabel("UMAP 2")
    plt.grid(True)

    # CriaÃ§Ã£o das pastas de saÃ­da por cluster
    os.makedirs(pasta_resultado, exist_ok=True)
    for i in range(12):
        os.makedirs(os.path.join(pasta_resultado, f"cluster_{i}"), exist_ok=True)

    # CÃ³pia das imagens para as respectivas pastas
    for nome_arquivo, label in zip(nomes, labels):
        origem = os.path.join(pasta_segmentadas, nome_arquivo)
        destino = os.path.join(pasta_resultado, f"cluster_{label}", nome_arquivo)
        if os.path.exists(origem):
            shutil.copy2(origem, destino)

    # Salva modelos
    os.makedirs(pasta_modelos, exist_ok=True)
    joblib.dump(scaler, os.path.join(pasta_modelos, f"scaler_{prefixo_modelo}.pkl"))
    joblib.dump(reducer, os.path.join(pasta_modelos, f"umap_{prefixo_modelo}.pkl"))
    joblib.dump(kmeans, os.path.join(pasta_modelos, f"kmeans_{prefixo_modelo}.pkl"))

    print(f"Modelos {prefixo_modelo} salvos em: {pasta_modelos}")

# Pasta onde salvar os modelos
PASTA_MODELOS = "../modelos_salvos"

# Cria figura com dois subplots
plt.figure(figsize=(16, 6))

# Processa pele
processar_e_plotar_salvar(
    csv_path="../features_pele.csv",
    titulo="ClusterizaÃ§Ã£o por Pele",
    subplot_index=1,
    pasta_segmentadas="../imagens_segmentadas",
    pasta_resultado="../resultado_pele",
    pasta_modelos=PASTA_MODELOS,
    prefixo_modelo="pele"
)

# Processa rosto
processar_e_plotar_salvar(
    csv_path="../features_rosto.csv",
    titulo="ClusterizaÃ§Ã£o por Rosto",
    subplot_index=2,
    pasta_segmentadas="../rostos_segmentados",
    pasta_resultado="../resultado_rosto",
    pasta_modelos=PASTA_MODELOS,
    prefixo_modelo="rosto"
)

# Legenda fora dos subplots
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
